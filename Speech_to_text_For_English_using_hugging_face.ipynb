```python
from transformers import pipeline
```

```python
p = pipeline("automatic-speech-recognition")
```

```python
import gradio as gr
```

```python
def transcribe(audio, state=""):
    text = p(audio)["text"]
    state += text + " "
    return state
```

```python

gr.Interface(
    fn=transcribe,
    inputs=[
        gr.Audio(sources=["microphone"], type="filepath")
    ],
    outputs=[
        "textbox"
    ],
    live=True).launch()
```
